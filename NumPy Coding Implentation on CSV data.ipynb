Given file "terrorismData.csv"

### Killed in USA 

Problem Statement :
Find value of killed column only where country == ‘United States’?
Print 0 in place of missing values.
Print count of Killed as integer value.
Output Format :

Killed1

Killed2

Killed3

Killed4

```python
import numpy as np
import csv

data=open ("terrorismData.csv",'r',encoding = "utf-8")
file_data=csv.DictReader(data,delimiter=',', skipinitialspace=True)

    
killed = []    # initiate empty list

for row in file_data:
    if row['Country']=='United States':
        killed.append(row['Killed'])
        
        
np_killed = np.array(killed)      # initiate numpy array using list

np_killed[np_killed==''] = 0

np_killed = np.array(np_killed, dtype = float)   # chage value to floar



for i in np_killed :
    print(int(i))
```

### Terror Casualty Attack

Problem Statement :
As we knew the Kargil ( in Jammu and Kashmir) War that took place between May 1999 and July 1999 (3 Months) ,so there was a huge conflict in Kashmir Valley during this period.
In this dataset, there is no information regarding the war between the two countries to find out the casualty during the war.
So find out the attack in this period in which maximum casualties happened.
Print the count of casualties (as integer), city in which that attack happened and name of attack group.
Note : Casualty = Killed + Wounded.Fill the empty value in killed or wounded feature to 0.
Output Format :

Casualty City TerroristGroup


```python
import csv
import numpy as np
with open("terrorismData.csv", encoding='utf8') as file_obj:
    file_data=csv.DictReader(file_obj)

    country=[]        # intiate empty lists
    kill=[]
    wound=[]
    group=[]
    
    
    for row in file_data:
        if (row["Year"]=="1999"):
            if (row["Country"]=="India" and row["State"]=="Jammu and Kashmir"):
                if (int(row["Month"])>=5 and int(row["Month"])<8):
                    if (int(row["Day"])>0):
                        
                        country.append(row["City"])       # update country and group list one by one
                        group.append(row["Group"])
                        
                        if row["Killed"]=="":             #anamalous value handled and list updated
                            kill.append("0.0")
                        else:
                            kill.append(row["Killed"])
                       
                    
                        if row["Wounded"]=="":
                            wound.append("0.0")
                        else:
                            wound.append(row["Wounded"])
                            
                            
    country=np.array(country)                 # convert the all list in a numpy array
    kill=np.array(kill,dtype=float)
    wound=np.array(wound,dtype=float)
    group=np.array(group)
    
    
    casulity_count=kill+wound
    max_val=np.argmax(casulity_count)
    print(int(casulity_count[max_val]),country[max_val],group[max_val])
```

### Terror Casualty Red 

Problem Statement :
Find the casualty in the Red Corridor States ? Mainly Red corridor states include Jharkhand, Odisha, Andhra Pradesh, and Chhattisgarh.
Note: Casualty=Killed +Wounded
Print count of Casualty as integer value.
Output Format :

Count


```python
import numpy as np
import csv

data=open ("terrorismData.csv",'r',encoding = "utf-8")
file_data=csv.DictReader(data,delimiter=',', skipinitialspace=True)

killed  =[]
wounded = []    # initiate empty list

for row in file_data:

    if (row['State']=='Jharkhand' or row['State']=='Odisha' or row['State']=='Andhra Pradesh' or row['State']=='Chhattisgarh'):

        if row["Killed"]=="":             #anamalous value handled and list updated
            killed.append("0.0")
        else:
            killed.append(row["Killed"])

        if row["Wounded"]=="":
            wounded.append("0.0")
        else:
            wounded.append(row["Wounded"])


killed = np.array(killed,dtype=float)      # convert in numpy list
wounded = np.array(wounded,dtype=float)

killed_wounded=killed+wounded


val=np.sum(killed_wounded)
print(int(val))
```

### Terror City Casualty

Problem Statement :
Find top 5 Indian Cities which has most number of casualties ?
Print top 5 cities along with total casualties in that city. Print count of Casualty as integer value.
Note: Ignoring the City which is Unknown.
Casualty = Killed + Wounded.
Output Format :

city_1 casualty_1

city_2 casualty_2

city_3 casualty_3

city_4 casualty_4

city_5 casualty_5



```python
import numpy as np
import csv

data=open ("terrorismData.csv",'r',encoding = "utf-8")
file_data=csv.DictReader(data,delimiter=',', skipinitialspace=True)

city_dict={}     # initiate empty dictionary

for row in file_data:
    if row["Country"]=='India':

            if row["Killed"]=="":             #anamalous value handled and list updated
                row["Killed"]=0
            else:
                row["Killed"]=int(float(row["Killed"]))

            if row["Wounded"]=="":
                row["Wounded"]=0
            else:
                row["Wounded"]=int(float(row["Wounded"]))


            key=row["City"]
            value=row['Killed']+row['Wounded']

            if key!='Unknown':
                if key in city_dict:
                    city_dict[key]+=value
                else:
                    city_dict[key]=value
            
citylist=sorted(city_dict, key=city_dict.get, reverse=True)[:5]
for i in citylist:
    print(i,city_dict[i])
```

### Terrorism FrequentDay 

Problem Statement :
Find the most frequent day of attack in a terrorismDataset ?
Note: Here np.unique can be used.
Print count of frequent day and number of attack as Integer value.
Output Format :

Day NumberOFAttack


```python
import numpy as np
import csv

data=open ("terrorismData.csv",'r',encoding = "utf-8")
file_data=csv.DictReader(data,delimiter=',', skipinitialspace=True)

day=[]
for row in file_data:
    day.append(row['Day'])
    
np_day=np.array(day,dtype=int)
day, count=np.unique(np_day,return_counts=True)

indexofmax=count.argmax()
print(day[indexofmax],count[indexofmax])
```
